### Gerando amostra aleatoria de dados

```python
import pandas as pd
```

```python
pd.set_option('display.max_colwidth', 250)
```

```python
df = pd.read_csv('./dataset_treino', index_col=[0])
```

```python
df
```

```python
dfSamples = df.sample(200)
```

```python
dfSamples
```

```python
#Divindo o data frame em duas partes
a = dfSamples.sample(frac=0.5,random_state=1000)
b= dfSamples.drop(a.index)
c = a.sample(frac=0.5,random_state=1000)
d = a.drop(c.index)
```

```python
dfSamples.to_csv('./tweets_treino')
a.to_csv('./adriano')
b.to_csv('./jefferson')
c.to_csv('./bruno')
d.to_csv('./matheus')
```
