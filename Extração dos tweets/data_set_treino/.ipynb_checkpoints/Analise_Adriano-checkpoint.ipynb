```python
dataframe_final = pd.read_csv('./dataset_treino_finalizado')
```

```python
def cleanText(text):
   return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|('https?:\/\/\S+')|(RT\S+)", " ", text).split())
```

```python
dataframe_final['tweet'] = dataframe_final['tweet'].apply(cleanText)
```

```python
df_limpo = dataframe_final['tweet']
```

```python
df_limpo.to_csv('./tweets_limpados')
```
