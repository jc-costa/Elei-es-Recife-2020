```python
import pandas as pd
from PIL import Image
from wordcloud import WordCloud, STOPWORDS
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/df_segundo_turno_final_sem_stopwords')
df.drop(columns={'Unnamed: 0'}, inplace=True)
```

```python
df['palavra_chave'].unique()
```

```python
MYmask = np.array(Image.open('./assets/imgs/estrela2.jpg'))
```

```python
stop_words = ["ou", "e", "+", "https"] + list(STOPWORDS)
stringAux = ' '.join(map(str, df.loc[(df['palavra_chave'] == 'Marília Arraes') & (df['classificacao'] == 'Negativo'), 'tweet'].values))
```

```python
wc = WordCloud(width = 500, background_color='white', stopwords = stop_words, height = 500, random_state = 21, max_font_size = 199, mask=MYmask, contour_color= 'steelblue').generate(stringAux)
```

```python
plt.imshow(wc, interpolation = "bilinear")
plt.axis('off')
plt.show()
```
